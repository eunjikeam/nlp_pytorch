## TF-IDF

```python
import pandas as pd
from math import log
```

```python
docs = [
        'drink cold milk',
        'drink cold water',
        'drink cold cola',
        'drink sweet juice',
        'drink sweet cola',
        'eat delicious bacon',
        'eat sweet mango',
        'eat delicious cherry',
        'eat sweet apple',
        'juice with sugar',
        'cola with sugar',
        'mango is fruit',
        'apple is fruit',
        'cherry is fruit',
        'Berlin is Germany',
        'Boston is USA',
        'Mercedes from Germany',
        'Mercedes is car',
        'Ford from USA',
        'Ford is car'
    ]
```

```python
vocab = list(set(w for doc in docs for w in doc.split())) # vocabulary
vocab.sort()
```

```python
N = len(docs)

def tf(t,d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t,d):
    return tf(t,d) * idf(t)
```

```python
## tf dataframe
result = []
for i, d in enumerate(docs):
    result.append([])
    for j, t in enumerate(vocab):
        result[-1].append(tf(t, d))
        
tf_ = pd.DataFrame(result, columns = vocab)
tf_.head()
```

```python
## idf dataframe
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))
    
idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
idf_
```

```python
result = []
for i, d in enumerate(docs):
    result.append([])
    for j, t in enumerate(vocab):
        result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_.head(5)
```
